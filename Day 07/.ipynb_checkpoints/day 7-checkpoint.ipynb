```python
#! /usr/bin/python
import re
import numpy
import numbers
```

```python
def create_gate( gate ):
    instruction, wire = gate[0]
    if not any(x in instruction for x in ['OR', 'AND', 'LSHIFT', 'RSHIFT', 'NOT']):
        instruct = [wire, [" ", "WIRE", instruction.split(' ')[0]]]
        try:
            instruct[1][2] = numpy.uint16(int(instruct[1][2]))
        except ValueError:
            print("create_gate: oops")
        if isinstance( instruct[1][2], numpy.uint16 ):
            instruct[1][1] = "SOURCE"
        print (instruct)
        gates.append(instruct)
    if any(x in instruction for x in ['OR', 'AND']):
        instruct = [wire, instruction.split(' ')]
        gates.append(instruct)
    if any(x in instruction for x in ['LSHIFT', 'RSHIFT']):
        instruct = [wire, instruction.split(' ')]
        gates.append(instruct)
    if any(x in instruction for x in ['NOT']):
        instruct = [wire, [" ", instruction.split(' ')[0], instruction.split(' ')[1]]]
        gates.append(instruct)
```

```python
def create_wire( wire, value ):
    wires[wire] = numpy.uint16(value)
    print ("create_wire: wire ", wire, " created")
    print (wire, ": ", value)

           
```

```python
match_source = re.compile('(^[0-9]*)')
match_create_gate = re.compile('(^[a-zA-Z0-9 ]*) -> ([a-z]*)')


filename = './input' #first star
filename = './input2' #second star
wires = {}
gates = []

print ("test")
#    READ ALL GATES
with open(filename) as f:
    for linen in f:
        line = linen.rstrip('\n')
        if not (match_create_gate.findall(line) == []):
            create_gate (match_create_gate.findall(line))
    
#    SETUP ALL SOURCES
create_wire('1', 1)
delgates = []
for gate in gates:
    if gate[1][1] == "SOURCE":
        create_wire(gate[0][0], gate[1][2])
        print("setup", gate)
        delgates.append(gate)

for gate in delgates:
        gates.remove(gate)

print ("-----------")


#    FIND GATES WITH COMPLETE SOURCES AND EXECUTE
while not (gates == []):
    print ("# gates left: ", len(gates))
    exec_gates = []
    for gate in gates:
        if ((gate[1][1]) == 'AND'):
            if (gate[1][0] in wires) and (gate[1][2] in wires):
                exec_gate = gate
                print ("SELECTING CANDIDATE GATE: ", exec_gate)
                print (exec_gate[1][0], ": ", wires[exec_gate[1][0]])
                print (exec_gate[1][2], ": ", wires[exec_gate[1][2]])
                gates.remove(exec_gate)
                break
        if (gate[1][1]) == 'OR':
            if (gate[1][0] in wires) and (gate[1][2] in wires):
                exec_gate = gate
                print ("SELECTING CANDIDATE GATE: ", exec_gate)
                print (exec_gate[1][0], ": ", wires[exec_gate[1][0]])
                print (exec_gate[1][2], ": ", wires[exec_gate[1][2]])
                gates.remove(exec_gate)
                break
        if (gate[1][1]) == 'LSHIFT':
            if (gate[1][0] in wires):
                exec_gate = gate
                print ("SELECTING CANDIDATE GATE: ", exec_gate)
                print (exec_gate[1][0], ": ", wires[exec_gate[1][0]])
                gates.remove(exec_gate)
                break            
        if (gate[1][1]) == 'RSHIFT':
            if (gate[1][0] in wires):
                exec_gate = gate
                print ("SELECTING CANDIDATE GATE: ", exec_gate)
                print (exec_gate[1][0], ": ", wires[exec_gate[1][0]])
                gates.remove(exec_gate)
                break            
        if (gate[1][1]) == 'NOT':
            if (gate[1][2] in wires):
                exec_gate = gate
                print ("SELECTING CANDIDATE GATE: ", exec_gate)
                print (exec_gate[1][2], ": ", wires[exec_gate[1][2]])
                gates.remove(exec_gate)
                break
        if (gate[1][1]) == 'WIRE':
            if (gate[1][2] in wires):
                exec_gate = gate
                print ("SELECTING CANDIDATE GATE: ", exec_gate)
                print (exec_gate[1][2], ": ", wires[exec_gate[1][2]])
                gates.remove(exec_gate)
                break

            

#   EXECUTE GATE
    print ("EXECUTING CANDIDATE GATE: ", exec_gate)

    if (exec_gate[1][1]) == 'AND':
        create_wire(exec_gate[0], numpy.uint16(wires[exec_gate[1][0]]) & numpy.uint16(wires[exec_gate[1][2]]))
    if (exec_gate[1][1]) == 'OR':
        create_wire(exec_gate[0], numpy.uint16(wires[exec_gate[1][0]]) | numpy.uint16(wires[exec_gate[1][2]]))
    if (exec_gate[1][1]) == 'LSHIFT':
        create_wire(exec_gate[0], numpy.uint16(wires[exec_gate[1][0]]) << int(exec_gate[1][2]))
    if (exec_gate[1][1]) == 'RSHIFT':
        create_wire(exec_gate[0], numpy.uint16(wires[exec_gate[1][0]]) >> int(exec_gate[1][2]))
    if (exec_gate[1][1]) == 'NOT':
        create_wire(exec_gate[0], ~numpy.uint16(wires[exec_gate[1][2]]))
    if (exec_gate[1][1]) == 'WIRE':
        create_wire(exec_gate[0], numpy.uint16(wires[exec_gate[1][2]]))
    print ("---")










```
